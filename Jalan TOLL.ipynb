```python
def find_pdf_in_html(self, html):
        """Find pdf url and title in a scihub html

        params:
            html: (str) scihub html in string format.

        returns:
            (dict) {
                'pdf_url': (str) real url of the pdf.
                'title': (str) title of the article.
            }
        """
        pdf = {}
        soup = BeautifulSoup(html, "html.parser")
        pdf_url = soup.find("embed", {"id": "pdf"}).attrs["src"].split("#")[0]
        pdf["pdf_url"] = (
            pdf_url.replace("https", "http").replace("zero","moscow") if "http" in pdf_url else "http:" + pdf_url
        )

        title = " ".join(self._trim(soup.title.text.split("|")[1]).split("/")).split(
            "."
        )[0]
        title = title if title else pdf["pdf_url"].split("/")[-1].split(".pdf")[0]
        pdf["title"] = self.check_title(title)
        print(
            STD_INFO
            + colored("PDF url", attrs=["bold"])
            + " -> \n\t%s" % (pdf["pdf_url"])
        )
        print(
            STD_INFO
            + colored("Article title", attrs=["bold"])
            + " -> \n\t%s" % (pdf["title"])
        )
        return pdf
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
# import org.openqa.selenium.support.ui.Select
from selenium.webdriver.support.ui import Select
import time
import json

Toll_list = []

jsondata = {}
agent={}

count = 0

#? webDriver for google chrome change path accordingly
driver = webdriver.Chrome("chromedriver.exe")
driver.get("https://bpjt.pu.go.id/cek-tarif-tol")

#?creating list for all toll road in Indonesia
select = Select(driver.find_element(By.NAME,'troad'))
time.sleep(2)
for i in select.options :
    Toll_list.append(i.text)
Toll_list.pop(0)

# Toll_list = ['Jakarta-Cikampek', 'Padalarang - Cileunyi']

z = 0
for toll in Toll_list :
    content = {}
    gates_in = []
    gates_out = []
    
    jsondata[Toll_list[z]]= []
       
    select.select_by_visible_text(toll)
    
    # ?creating list for gates in and gates out
    select2 = Select(driver.find_element(By.NAME,'tgates'))
    time.sleep(2)
    for j in select2.options :
        gates_in.append(j.text)
    gates_in.pop(0)
    
    n=0
    for x in gates_in :
        content[n] = x
        n += 1
        
    # print ("start")
    # print (content)
    # print ("end")
    
    select3 = Select(driver.find_element(By.NAME,'tgates2'))
    time.sleep(2)
    for k in select3.options :
        gates_out.append(k.text)
    gates_out.pop(0)
    
    #?creating dictionary
    # jsondata[Toll_list[z]] = content
    # print (type(jsondata[Toll_list[z]]))
    # print (type(content))
    jsondata[Toll_list[z]].append (content)
    z += 1

    # print("data processing")
    # print ("---------------------------------------------------------------------")

print(json.dumps(jsondata))
# Serializing json
json_object = json.dumps(jsondata, indent=4)
# Writing to sample.json
with open("sample.json", "w") as outfile:
    outfile.write(json_object)
    
driver.close()
```
