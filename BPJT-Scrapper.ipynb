```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time
import json

Toll_list = []

jsondata = {}
agent={}

count = 0

#? webDriver for google chrome change path accordingly
driver = webdriver.Chrome("chromedriver.exe")
driver.get("https://bpjt.pu.go.id/cek-tarif-tol")

#?creating list for all toll road in Indonesia
select = Select(driver.find_element(By.NAME,'troad'))
time.sleep(2)
for i in select.options :
    Toll_list.append(i.text)
Toll_list.pop(0)

z = 0
for toll in Toll_list :
    content = {}
    gates_in = []
    gates_out = []
    
    jsondata[Toll_list[z]]= []
       
    select.select_by_visible_text(toll)
    
    # ?creating list for gates in and gates out
    select2 = Select(driver.find_element(By.NAME,'tgates'))
    time.sleep(2)
    for j in select2.options :
        gates_in.append(j.text)
    gates_in.pop(0)
    
    n=0
    for x in gates_in :
        content[n] = x
        n += 1
    
    select3 = Select(driver.find_element(By.NAME,'tgates2'))
    time.sleep(2)
    for k in select3.options :
        gates_out.append(k.text)
    gates_out.pop(0)
    
    #?creating dictionary
    jsondata[Toll_list[z]].append (content)
    z += 1

print(json.dumps(jsondata))
# Serializing json
json_object = json.dumps(jsondata, indent=4)
# Writing to sample.json
with open("sample.json", "w") as outfile:
    outfile.write(json_object)
    
driver.close()
```
